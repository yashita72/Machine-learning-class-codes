```python
import pandas as pd
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
titanic=pd.read_csv('/kaggle/input/datasets/aliaamiri/titanic-passengers-and-crew-complete-list/titanic.csv')
titanic.head()
```

# *1. SCATTER PLOT (Numerical-numerical data)*

```python
sns.scatterplot(x=tips['tip'],y=tips['total_bill'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# *2. BARPLOT (Numerical-Categorical data)*

```python
sns.barplot(x=titanic['embarked'],y=titanic['fare'],hue=titanic['gender'])
```

# *3. BOXPLOT (Numerical-Categorical data)*

```python
sns.boxplot(x=titanic['gender'],y=titanic['age'])
```

# *4. DISTPLOT (Numerical-Categorical data)*

```python
sns.distplot(titanic[titanic['survived']=='no']['age'],hist=False)
sns.distplot(titanic[titanic['survived']=='yes']['age'],hist=False)
```

# *5. HEATMAP (Categorical-Categorical data)*

```python
cross=pd.crosstab(titanic['survived'],titanic['gender'])
```

```python
cross
```

```python
sns.heatmap(cross)
```

# *6. CLUSTER (Categorical-Categorical data)*

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['sibsp'],titanic['survived']))
```

# *7. PAIRPLOT (Numerical-Numerical data)*

```python
sns.pairplot(iris,hue='species')
```

# *8. PLOTLINE (Numerical-Numerical data)*

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
new
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```
