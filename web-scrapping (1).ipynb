```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
for i in range(1,501):
    response=requests.get("https://www.ambitionbox.com/list-of-companies?page={}".format(i),headers=headers).text
```

```python
response
```

```python
soup=BeautifulSoup(response ,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in range(1,21):
    print(soup.find_all('h2')[i].text.strip())
```

```python
for i in soup.find_all('div',class_="rating_text"):
    print(i.text.strip())
```

```python
for i in soup.find_all('span',class_="companyCardWrapper__companyRatingCount"):
    print(i.text.strip())
```

```python
for i in soup.find_all('span' ,class_="companyCardWrapper__ActionCount"):
    print(i)
len(soup.find_all('span' ,class_="companyCardWrapper__ActionCount"))
```

```python
company=soup.find_all('div',class_="companyCardWrapper")
```

```python
len(company)
```

```python
name=[]
rating=[]
review_count=[]
interlinking=[]
salaries=[]
interviews=[]
jobs=[]
benefits=[]
```

```python
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_="rating_text").text.strip())
    interlinking.append(i.find('span',class_="companyCardWrapper__interLinking").text.strip())
    spans = i.find_all('span', class_='companyCardWrapper__ActionCount')
    review_count.append(spans[0].text.strip())
    salaries.append(spans[1].text.strip())
    interviews.append(spans[2].text.strip())
    jobs.append(spans[3].text.strip())
    benefits.append(spans[4].text.strip())
```
