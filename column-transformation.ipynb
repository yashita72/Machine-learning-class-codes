```python
import pandas as pd
```

```python
df=pd.read_csv('/kaggle/input/datasets/divjotbedi/customer-dataset-with-education-age-gender/customer.csv')
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X,test_X,train_y,test_y=train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.2,random_state=42)
```

```python
train_X
```

```python
test_X
```

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
                              ('tnf1',OrdinalEncoder(categories=[['Poor', 'Average', 'Good', 'Excellent'],['School', 'UG', 'PG']]), ['review', 'education']),
                              ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['gender'])],
                             remainder='passthrough')
```

```python
train_X_tranformed=transformer.fit_transform(train_X)
train_X_tranformed=pd.DataFrame(train_X_tranformed,columns=['review','education','female','age'])
```

```python
test_X_tranformed=transformer.fit_transform(test_X)
test_X_tranformed=pd.DataFrame(test_X_tranformed,columns=['review','education','female','age'])
```
