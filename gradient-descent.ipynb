```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/datasets/yashitagaur/placement-data/placement.csv')
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
m = 0
b = 0
lr = 0.01
```

```python
epochs = 1000

n = len(X)
```

```python
for i in range(epochs):

    y_pred = m*X + b     # prediction

    # gradients (derivatives)
    dm = (-2/n) * sum(X * (y - y_pred))
    db = (-2/n) * sum(y - y_pred)

    # update rule
    m = m - lr * dm
    b = b - lr * db
```

```python
print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
# example cgpa = 8.5
x_new = 8.5
prediction = m*x_new + b

print("Predicted package:", prediction)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, m*X + b)
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```
